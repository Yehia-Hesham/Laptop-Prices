# Data Description

- Company         : Company that sells the laptop
- Product         : Name of the Laptop
- TypeName        : Type of Laptop (ex: Ultrabook, Notebook, Gaming or workstation) 
- Inches          : size of the laptop screen/body
- ScreenResolution: the Type,size and technology of the display
- Cpu             : central processing unit (Compute unit) type
- Ram             : Amount of Random Access Memory available to the system
- Memory          : Type of Storage available for the system (ex: Flash, HDD, SSD or Hybrid)
- Gpu             : full Name of the Graphical Processing Unit
- OpSys           : Operating System
- Weight          : weight of the laptop in Kg
- Price_euros     : Price of the laptop in euros

# Data Wrangling / Cleaning example

```python
example_df = df_laptops.iloc[:5,:]
example_df
```

```python
#preprocess Company column
keys = ['Mediacom', 'Microsoft', 'Razer', 'Google'] + ['Xiaomi','Chuwi','Huawei'] + ['Vero','Fujitsu']
values = ['other (european/American)'] * 4 + ['other (Chinease/Korean)'] * 3 + ['other'] * 2
replacement_dict = dict(zip(keys, values))
replacement_dict
```

```python
example_df.loc[:, 'Company'].replace(replacement_dict, inplace = True)
example_df
```

```python
categories = ['Apple', 'HP', 'Acer', 'Asus', 'Dell', 'Lenovo',
              'other (Chinease/Korean)', 'MSI', 'other (european/American)',
              'Toshiba', 'Samsung', 'LG']
example_df[[f'Company: {x}' for x in categories]] = example_df['Company'].apply(
    lambda x: pd.Series([1 if y == x else 0 for y in categories]))
example_df.drop('Company', axis = 1, inplace = True)
example_df
```

```python
#preprocess TypeName column
example_df.loc[:, 'TypeName'].replace({'Netbook': 'Notebook'}, inplace = True)
categories = ['Ultrabook', 'Gaming', '2 in 1 Convertible', 'Workstation']
example_df[[f'Type: {x}' for x in categories]] = example_df['TypeName'].apply(
    lambda x: pd.Series([1 if y == x else 0 for y in categories]))
example_df.drop('TypeName', axis = 1, inplace = True)
example_df
```

```python
import re
#preprocessing Memory column
categories = ['SSD', 'HDD', 'Flash', 'Hybrid']
example_df[[f'Memory: {x}' for x in categories]] = example_df['Memory'].apply(lambda x: pd.Series(
    [1 if y in x else 0 for y in categories]))
example_df[['Memory (GB): Primary', 'Memory (GB): Secondary']] = example_df['Memory'].apply(
    lambda x: pd.Series([int(y.replace('TB', '')) * 1024 if 'TB' in y else int(y.replace('GB', '')) for y in (re.findall(r'-?\d+\w+\.?\d*\w+', x) if len(re.findall(r'-?\d+\w+\.?\d*\w+', x)) == 2 else re.findall(r'-?\d+\w+\.?\d*\w+', x) + ['0GB'])])).fillna(0).astype('int')
example_df.drop('Memory', axis = 1, inplace = True)
example_df
```

```python
#preprocess Ram column
example_df['Ram Size (GB)'] = example_df['Ram'].str.replace('GB', '').astype('int')
example_df.drop('Ram', axis = 1, inplace = True)
example_df
```

```python
#preprocess Weight column
example_df['Weight (kg)'] = example_df['Weight'].str.replace('kg', '').astype('float16')
example_df.drop('Weight', axis = 1, inplace = True)
example_df
```

```python
#preprocess OpSys column
example_df.loc[:, 'OpSys'].replace({'Mac OS X': 'macOS', 'Windows 10 S': 'Windows 10'}, inplace = True)
categories = ['macOS', 'Windows 10', 'Linux', 'Android', 'Chrome OS', 'Windows 7']
example_df[[f'OpSys: {x}' for x in categories]] = example_df['OpSys'].apply(lambda x: pd.Series([1 if y == x else 0 for y in categories]))
example_df.drop('OpSys', axis = 1, inplace = True)
example_df
```

```python
#preprocess Inches column
example_df['Screen Size (inch)'] = example_df['Inches']
example_df.drop('Inches', axis = 1, inplace = True)
example_df
```

```python
#preprocess ScreenResolution column
screen_features = ['IPS', 'Touchscreen', 'Retina Display']

example_df[['Screen Resolution (px): Horizontal', 
    'Screen Resolution (px): Vertical', 
    'Screen Features: IPS', 
    'Screen Features: Touchscreen', 
    'Screen Features: Retina Display']] = example_df['ScreenResolution'].apply(
    lambda x: pd.Series([int(y) for y in x.split()[-1].split('x')] + [1 if y in x else 0 for y in screen_features]))
example_df.drop(['ScreenResolution'], axis = 1, inplace = True)
example_df
```

```python
import difflib

#preprocess CPUs and GPUs
example_df['gpu_name'] = [name + (50 -len(name)) * " " for name in example_df.Gpu]
example_df['cpu_name'] = [name + (50 -len(name)) * " " for name in example_df.Cpu]

gpu_df['Videocard Name'] = [name + (50 -len(name)) * " " for name in gpu_df['Videocard Name']]
cpu_df['CPU Name'] =       [name + (50 -len(name)) * " " for name in cpu_df['CPU Name']]

for idx_laptop, row_laptop in example_df.iterrows():  
        gpu_key = difflib.get_close_matches(example_df.loc[idx_laptop,'gpu_name'], gpu_df['Videocard Name'] ,n=1)[0]
        example_df.loc[idx_laptop,'gpu_name_key'] = gpu_key

        cpu_key = difflib.get_close_matches(example_df.loc[idx_laptop,'cpu_name'], cpu_df['CPU Name'] ,n=1)[0]
        example_df.loc[idx_laptop,'cpu_name_key'] = cpu_key
example_df
```

```python
example_df[['cpu_name','cpu_name_key','gpu_name','gpu_name_key']]
```

```python
example_df['gpu_G3D_Mark'] = 0
example_df['gpu_G2D_Mark'] = 0
example_df['cpu_Thread_Mark'] = 0
example_df['cpu_mark'] = 0
example_df['cpu_cores'] = 0

gpu_df['G3D Mark -']    = gpu_df['G3D Mark -'].str.replace(",","")
gpu_df['G2D Mark -']    = gpu_df['G2D Mark -'].str.replace(",","")
cpu_df['Cores']         = cpu_df['Cores']
cpu_df['CPU Mark -']    = cpu_df['CPU Mark -'].str.replace(",","")
cpu_df['Thread Mark -'] = cpu_df['Thread Mark -'].str.replace(",","")

for idx_laptop, row_laptop in example_df.iterrows():

    gpu_key = example_df.loc[idx_laptop, 'gpu_name_key'] 
    example_df.loc[idx_laptop,'gpu_G3D_Mark'] = gpu_df.query("{0} == @gpu_key".format('`Videocard Name`')).iloc[0]['G3D Mark -'].replace(",","")
    example_df.loc[idx_laptop,'gpu_G2D_Mark'] = gpu_df.query("{0} == @gpu_key".format('`Videocard Name`')).iloc[0]['G2D Mark -'].replace(",","")

    cpu_key = example_df.loc[idx_laptop,'cpu_name_key'] 
    example_df.loc[idx_laptop,'cpu_Thread_Mark'] = cpu_df.query("{0} == @cpu_key".format('`CPU Name`')).iloc[0]['Thread Mark -'].replace(",","")
    example_df.loc[idx_laptop,'cpu_mark'] = cpu_df.query("{0} == @cpu_key".format('`CPU Name`')).iloc[0]['CPU Mark -'].replace(",","")
    example_df.loc[idx_laptop,'cpu_cores'] = cpu_df.query("{0} == @cpu_key".format('`CPU Name`')).iloc[0]['Cores']


example_df['gpu_G3D_Mark'] = example_df['gpu_G3D_Mark'].astype('int')
example_df['gpu_G2D_Mark'] = example_df['gpu_G2D_Mark'].astype('int')
example_df['cpu_Thread_Mark'] = example_df['cpu_Thread_Mark'].astype('int')
example_df['cpu_mark'] = example_df['cpu_mark'].astype('int')

#preprocess CPU and GPU columns
example_df.drop(['Cpu', 'Gpu','cpu_name','gpu_name','cpu_name_key','gpu_name_key'], axis = 1, inplace = True)
example_df
```
