<a id="Question1"></a>
# according to my price bracket, what storage configuration should I use?

```python
sql = """
SELECT price_bracket, l.memories, COUNT(l.price_bracket)
FROM laptops l
GROUP BY price_bracket,memories
ORDER BY price_bracket,memories"""
```

```python
cursor = connection.execute(sql)

cursor.fetchall()
```

```python
sql = """
SELECT l.price_bracket, l.memory_primary,
l.memory_secondary,
COUNT(*)
FROM laptops l
GROUP BY l.price_bracket,
    l.memory_primary,
    l.memory_secondary
ORDER BY l.price_bracket"""
```

```python
cursor = connection.execute(sql)

cursor.fetchall()
```

<a id="Question2"></a>
# for each price bracket, which companies use the strongest processors? (for programmers)

```python
sql = """
SELECT l.price_bracket, l.companies, AVG(l.cpu_mark)
FROM laptops l 
GROUP BY l.price_bracket, l.companies 
ORDER BY l.price_bracket, AVG(l.cpu_mark) DESC
"""
```

```python
cursor = connection.execute(sql)

cursor.fetchall()
```

<a id="Question3"></a>
# for each price bracket, which companies use the strongest Video Cards? (for editors)

```python
sql = """
SELECT l.price_bracket, l.companies, AVG(l.gpu_g2d_mark)
FROM laptops l 
GROUP BY l.price_bracket, l.companies 
ORDER BY l.price_bracket, AVG(l.gpu_g2d_mark) DESC
"""

cursor = connection.execute(sql)

cursor.fetchall()
```

    <li>For Budget Laptops, Acer Dell and Hp have the highest G2D/G3D Marks.</li>
    <li>For Mid Range Laptops, Msi beats out all of the competition with over 30% scores.</li>
    <li>For High End, Acer, Asus and Msi are above the competition by a respectable Margin.</li>
</ul>

```python
sql = """
SELECT l.price_bracket, l.companies, AVG(l.gpu_g3d_mark)
FROM laptops l 
GROUP BY l.price_bracket, l.companies 
ORDER BY l.price_bracket, AVG(l.gpu_g3d_mark) DESC
"""

cursor = connection.execute(sql)

cursor.fetchall()
```

# most popular CPUs for each Ram configuration?

```python
sql = """
SELECT ram_gb, cpu_name_key, COUNT(*)
FROM laptops
GROUP BY ram_gb, cpu_name
ORDER BY ram_gb, COUNT(*) DESC
"""

cursor = connection.execute(sql)

result = cursor.fetchall()

df_res = pd.DataFrame(result,columns=['ram_gb', 'cpu_name_key', 'COUNT(*)'])
df_res.sample(5)
```

```python
df_res.groupby('ram_gb').first()
```

# most popular GPUs for each screen resolution?

```python
sql = """
SELECT  gpu_name, gpu_g2d_mark, screen_resolution_w, screen_resolution_l, COUNT(*)
FROM laptops
GROUP BY gpu_name, screen_resolution_w, screen_resolution_l
ORDER BY gpu_name, screen_resolution_w, screen_resolution_l, COUNT(*) DESC
"""

cursor = connection.execute(sql)

result = cursor.fetchall()

df_res = pd.DataFrame(result,columns=['gpu_name','gpu_g2d_mark', 'screen_resolution_w', 'screen_resolution_l' , 'COUNT(*)'])
df_res.sample(5)
```

```python
df_res.groupby(['screen_resolution_w','screen_resolution_l']).first()
```

```python
sql = """
SELECT  gpu_name, gpu_g3d_mark, screen_resolution_w, screen_resolution_l, COUNT(*)
FROM laptops
GROUP BY gpu_name, screen_resolution_w, screen_resolution_l
ORDER BY gpu_name, screen_resolution_w, screen_resolution_l, COUNT(*) DESC
"""

cursor = connection.execute(sql)

result = cursor.fetchall()

df_res = pd.DataFrame(result,columns=['gpu_name','gpu_g3d_mark', 'screen_resolution_w', 'screen_resolution_l' , 'COUNT(*)'])
df_res.sample(5)
```

# are there any companies that overprice their products in comparison to the others?

```python
sql = """
SELECT  price_bracket, companies, COUNT(*) count, AVG(price_euros) mean
FROM laptops
GROUP BY price_bracket,companies
ORDER BY price_bracket,companies
"""

cursor = connection.execute(sql)

cursor.fetchall()
```

<a id="Question8"></a>
# for a display to be considered 'retina', what's the threshold for each screen size?

```python
sql = """
SELECT inches, screen_resolution_l, screen_resolution_w,
    ((screen_resolution_w * screen_resolution_w) + (screen_resolution_l * screen_resolution_l)) as PPI
FROM laptops
WHERE screens = 'Panel ips,Retina display,'
"""

cursor = connection.execute(sql)

result = cursor.fetchall()

df_res = pd.DataFrame(result,columns=['inches', 'screen_resolution_l', 'screen_resolution_w','PPI'])
df_res
```

```python
df_res['PPI'] = np.sqrt(df_res['PPI']) / df_res['inches']
```

<a id="Question9"></a>
# do touchscreens really affect overall laptop price?

```python
sql = """
SELECT COUNT(*), AVG(price_euros) 
FROM laptops
WHERE screens = 'Touchscreen,'
"""
cursor = connection.execute(sql)

cursor.fetchall()
```

```python
sql = """
SELECT COUNT(*), AVG(price_euros) 
FROM laptops
WHERE screens != 'Touchscreen,'
"""

cursor = connection.execute(sql)

cursor.fetchall()
```
